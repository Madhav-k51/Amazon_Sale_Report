```python
#---------plot representation--------
```

```python
df['month'] = df.Date.dt.to_period('M') 
df['revenue'] = df['Amount']
```

```python
df_group = df.groupby(['month', 'Category'])['revenue'].sum().reset_index()
df_pivot = df_group.pivot(index='month', columns='Category', values='revenue')
```

```python
df_group
```

```python
df_pivot
```

```python
months = df_pivot.index.strftime('%b') 

for col in df_pivot.columns:
    plt.plot(months, df_pivot[col], marker='o', label=col)

plt.title('Monthly Revenue Trend by Clothing Category (Line Plot)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Top 5 Orders
```

```python
df1.groupby('ship-state')['Order ID'].count().nlargest(5).reset_index(name = 'Top 5')
```

```python
q = [0, 0.3, 0.7, 1.0]
labels = ['Low', 'Medium', 'High']

df1['level'] = pd.qcut(df1['Total_Orders'], q=q, labels=labels, duplicates='drop')
```
