```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("Amazon_Sales_Cleaned_data.csv")
df
```

```python
df.drop(['month', 'revenue'], axis = 1, inplace = True)
```

```python
df.groupby('ship-state')['Amount'].agg(['sum', 'mean', 'median', 'std', 'count'])
```

```python
test1 = stats.ttest_ind(df.Amount, df['Order ID'].count())
z_stat, p_value = test1.statistic, test1.pvalue
print(f"t-statistic: {z_stat}")
print(f"P_value: {p_value}")
```

```python
df.Amount.mean()
```

```python
mu = df.Amount.mean() # Hypothesized mean
alpha = 0.05  # Significance level

test = stats.ttest_1samp(df.Amount, mu)
t_stat, p_value = test.statistic, test.pvalue

print(f"t-statistic: {t_stat:.6f}")
print(f"p-value: {p_value:.6f}")
print(f"df: {test.df}")
```
